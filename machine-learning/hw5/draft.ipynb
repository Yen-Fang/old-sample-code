```python
#Q17-Q20
import numpy as np
from libsvm.svmutil import *
train_y, train_x = svm_read_problem('train1.txt')
test_y, test_x = svm_read_problem('test1.txt')
ccc=0
train_x = [list(train_x[i].values()) for i in range(len(train_x)) if train_y[i] in [11, 26]]
test_x = [list(test_x[i].values()) for i in range(len(test_x)) if test_y[i] in [11, 26]]
train_y = [1 if x == 11 else -1 if x == 26 else None for x in train_y if x in [11, 26]]
test_y = [1 if x == 11 else -1 if x == 26 else None for x in test_y if x in [11, 26]]
X = np.array(train_x)
y = np.array(train_y)
Ein_gt = np.zeros(T)
T = 1000  # Number of iterations
N = X.shape[0]  # Number of data points
weights = np.ones(N) / N  # Initialize weights uniformly

alpha_t = np.zeros(T)
best_stump = {'feature': None, 'threshold': None, 'parity': None}
best_stump_lst = []

for t in range(T):
    min_err = float('inf')
    print(t)
    # Step 1: Sort features
    for i in range(X.shape[1]):
        sorted_indices = np.argsort(X[:, i])
        sorted_X = X[sorted_indices]
        #print(sorted_X[:, i])
        sorted_y = y[sorted_indices]
        sorted_weights = weights[sorted_indices]
        # Step 2: Find best threshold and parity
        for j in range(-1, N - 1):
            if j == -1:
                threshold = -99999999999
            else:
                threshold = (sorted_X[j, i] + sorted_X[j + 1, i]) / 2
            #print(threshold)
            for s in [-1, 1]:
                predictions = s * np.where(sorted_X[:, i] < threshold, 1, -1)
                err = np.sum(sorted_weights[sorted_y != predictions])

                if err < min_err:
                    min_err = err
                    best_stump['feature'] = i
                    best_stump['threshold'] = threshold
                    best_stump['parity'] = s
    #print(ccc, best_stump)
    ccc+=1
    # Step 3: Calculate alpha_t
    eps = min_err / np.sum(weights)
    alpha_t[t] = np.log(np.sqrt((1 - eps) / eps))


    # Update weights
    predictions = best_stump['parity'] * np.where(X[:, best_stump['feature']] < best_stump['threshold'], 1, -1)
    best_stump_lst.append((best_stump['parity'], best_stump['feature'], best_stump['threshold']))
    #print(sum(predictions))
    incorrect_mask = y != predictions  # Mask for incorrect predictions
    correct_mask = y == predictions  # Mask for correct predictions
    weights *= np.where(incorrect_mask, np.exp(alpha_t[t]), 1.0)  # Incorrect examples
    weights /= np.where(correct_mask, np.exp(alpha_t[t]), 1.0)  # Correct examples
#     print(len(weights))
    print(sum(weights))    
    weights /= np.sum(weights)  # Normalize weights
    
    error = 0
    best_stump['parity'] = best_stump_lst[t][0]
    best_stump['feature'] = best_stump_lst[t][1]
    best_stump['threshold'] = best_stump_lst[t][2]
    gt_pred = best_stump['parity'] * np.where(X[:, best_stump['feature']] < best_stump['threshold'], 1, -1)
    for i in range(len(y)):
        if gt_pred[i] != y[i]:
            error += 1
    print(error/len(y))
    Ein_gt[t] = error/len(y)
```
